```python
import pandas as pd
df = pd.DataFrame()

import requests
from bs4 import BeautifulSoup

URL = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
PostalCode = []
Borough = []
Neighborhood = []
res = requests.get(URL).text
soup = BeautifulSoup(res,'lxml')
for items in soup.find('table', class_='wikitable sortable').find_all('tr')[1::1]:
    data = items.find_all(['th','td'])
    try:
        pc = data[0].text
        Bor = data[1].text
        Neigh = data[2].text
        PostalCode.append(pc)
        Borough.append(Bor)
        Neighborhood.append(Neigh)
    except IndexError:pass
  
```

```python
import pandas as pd
df1 = pd.DataFrame()

df1['PostalCode'] = PostalCode
df1['Borough']= Borough
df1['Neighborhood'] = Neighborhood

df1
```
